```python
from data import load_dataset, denoise_data
from matplotlib import pyplot as plt
import numpy as np
```

```python
def mse(X_raw, X_denoise):
    return np.mean(np.sum((X_denoise - X_raw) ** 2, axis=1) / np.sum(X_raw ** 2, axis=1))

import numpy as np

def psnr(denoised_signal, original_signal, max_amplitude=None):
    """
    Calculates the Peak Signal-to-Noise Ratio (PSNR) for 1D signals like ECG.

    Args:
        
        denoised_signal (np.ndarray): The processed (denoised) signal = I
        original_signal (np.ndarray): The clean (ground truth) signal = K
        max_amplitude (float): The maximum possible amplitude of the signal.
                               If None, it uses the maximum value found in the original_signal.
                               For clinical ECG, this might be a fixed mV value.

    Returns:
        float: The PSNR value in decibels (dB).
    """
    original = np.asarray(original_signal)
    denoised = np.asarray(denoised_signal)
    
    if original.shape != denoised.shape:
        raise ValueError("Original and denoised signals must have the same shape.")

    mse = np.mean((original - denoised) ** 2)

    if mse == 0:
        return float('inf') 

    if max_amplitude is None:
        R = np.max(np.abs(original))
    else:
        R = max_amplitude

    psnr_value = 10 * np.log10(R**2 / mse)
    num = R**2
    denom = mse
    print(num, denom)
    return psnr_value

# Example Usage:
# Assume you have your clean_ecg and denoised_ecg numpy arrays
# psnr_score = psnr_ecg(clean_ecg, denoised_ecg, max_amplitude=1.0)


def experiment_denoise(X_raw):
    mse_list = []
    psnr_list = []
    atoms = range(8, 51)
    
    for n_atoms in atoms:
        print(n_atoms)
        
        X_denoise = denoise_data(X_raw, n_atoms, 0)
        mse_list.append(mse(X_raw, X_denoise))
        psnr_list.append(psnr(X_raw, X_denoise))

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    axs[0].plot(atoms, mse_list, color="red")
    axs[0].set_title("MSE")
    axs[0].set_xlabel("Number of atoms")
    axs[0].grid(True)
    
    axs[1].plot(atoms, psnr_list, color="blue")
    axs[1].set_title("PSNR")
    axs[1].set_xlabel("Number of atoms")
    axs[1].grid(True)

    plt.tight_layout()
```

```python
X_raw, _ = load_dataset("amazon", denoise=False)
experiment_denoise(X_raw)
plt.savefig("denoise_amazon.png", format="png", dpi=300)
plt.show()
```
