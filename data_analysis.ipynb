```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from typing import Union, List

def plot_price_window(TICKER,
    price_window: Union[np.ndarray, List[float]], 
    start_date_str: str, 
    window_length: int = 60
) -> None:
    """
    Plots a time series window (e.g., a 60-day price sequence) against its dates.

    Args:
        price_window: A 1D array or list containing the price data for the window.
                      (Expected length: window_length, e.g., 60)
        start_date_str: A string representing the start date of the window (e.g., '2024-01-01').
        window_length: The length of the time series window (default is 60).
    """
    
    if len(price_window) != window_length:
        print(f"Warning: The input array has length {len(price_window)}, but `window_length` is set to {window_length}.")

    try:
        start_date = datetime.strptime(start_date_str, '%Y-%m-%d')
    except ValueError:
        print("Error: Could not parse the start date string. Please use 'YYYY-MM-DD' format.")
        return

    dates = [start_date + timedelta(days=i) for i in range(window_length)]

    plt.figure(figsize=(12, 6))

    plt.plot(dates, price_window, marker='o', linestyle='-', color='#007ACC', linewidth=2, markersize=4)

    plt.title(f'{TICKER} Prices from {start_date_str} to {dates[-1].strftime("%Y-%m-%d")}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m-%d'))
    plt.xticks(rotation=45, ha='right')
    
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

START_DATE = '1997-05-16'

# 3. Call the function
plot_price_window("AMZN", X_0, START_DATE)
```
